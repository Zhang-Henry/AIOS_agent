```python
import matplotlib.pyplot as plt

# Data points for FNR and FPR
FNR = [0.008, 0.025538462, 0.073538462, 0.132538462, 0.175, 0.189, 0.204076923, 0.247615385, 0.319076923, 0.397538462, 0.518461538, 0.719769231, 0.853923077]
FPR = [0.997, 0.969692308, 0.889461538, 0.851461538, 0.818307692, 0.807307692, 0.793384615, 0.785384615, 0.539153846, 0.512153846, 0.383153846, 0.238307692, 0.063769231]
# Adjusting the font size for better visibility
plt.figure(figsize=(8, 6))
plt.plot(FNR, FPR, marker='o', linestyle='-', label='FPR vs FNR Curve')

# Adding grid lines every 0.1
plt.xticks([i * 0.1 for i in range(0, 11)], fontsize=12)
plt.yticks([i * 0.1 for i in range(0, 11)], fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Adding coordinates to each point with two decimal places, alternating left and right for overlapping points
for i, (fnr, fpr) in enumerate(zip(FNR, FPR)):
    # Adjusting text placement to ensure labels do not go outside the plot area
    if fnr < 0.05:
        plt.text(fnr + 0.03, fpr, f'({fnr:.2f}, {fpr:.2f})', fontsize=10, ha='left', va='center')
    elif i % 2 == 0:
        plt.text(fnr + 0.02, fpr, f'({fnr:.2f}, {fpr:.2f})', fontsize=10, ha='left', va='center')
    else:
        plt.text(fnr - 0.06, fpr, f'({fnr:.2f}, {fpr:.2f})', fontsize=10, ha='right', va='center')

# Setting larger labels and title
plt.xlabel('False Negative Rate (FNR)', fontsize=18)
plt.ylabel('False Positive Rate (FPR)', fontsize=18)
plt.legend(fontsize=18)

# Save the plot as a PDF file with larger font sizes
pdf_file_path_large_font = 'ppl.pdf'
plt.savefig(pdf_file_path_large_font, format='pdf')
plt.show()

pdf_file_path_large_font
```
